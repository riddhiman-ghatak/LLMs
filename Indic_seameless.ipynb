```python
from transformers import SeamlessM4Tv2Model, AutoProcessor
from transformers import AutoProcessor
```

```python
processor = AutoProcessor.from_pretrained("facebook/seamless-m4t-v2-large")
model = SeamlessM4Tv2Model.from_pretrained("facebook/seamless-m4t-v2-large")
```

```python
article_en = "In a big city, there was an old bookstore. The outside looked old, but inside, there were many interesting stories. The books had been there for a long time, and you could smell the paper and ink."
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
# process input
text_inputs = processor(article_en, src_lang="eng", return_tensors="pt").to(device)
```

```python
# generate translation
output_tokens = model.generate(**text_inputs, tgt_lang="ben", generate_speech=False)
translated_text_from_text = processor.decode(output_tokens[0].tolist()[0], skip_special_tokens=True)
print(f"Translation from text: {translated_text_from_text}")
```
