```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
article_en = "In a big city, there was an old bookstore. The outside looked old, but inside, there were many interesting stories. The books had been there for a long time, and you could smell the paper and ink."
```

```python
tokenizer.src_lang = "en_XX"
encoded_hi = tokenizer(article_en, return_tensors="pt")
generated_tokens = model.generate(
    **encoded_hi,
    forced_bos_token_id=tokenizer.lang_code_to_id["bn_IN"]
)
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```
